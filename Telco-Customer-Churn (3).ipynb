  <h2 style="font-size: 24px; font-weight: bold; color: #0077b6;">What is Customer Churn?</h2>
  <p>Customer churn, also known as customer attrition or customer turnover, refers to the rate at which customers stop doing business with a company over a certain period of time. In the context of the telecommunications industry, customer churn specifically refers to the percentage of subscribers or users who discontinue their services with a particular telecom provider. This could happen for a variety of reasons, such as dissatisfaction with the services provided, better offers from competitors, or a change in the customer's needs or preferences. Reducing customer churn is a key goal for telecom companies, as it can have a significant impact on their revenue and profitability.</p>

**1. Loading libraries and data**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE

# Ignore warnings
warnings.filterwarnings('ignore')
```

**3. preprocessing data**

```python
# Delet Unused Columns
df.drop(columns = ['Zip Code', 'Latitude', 'Longitude', 'City',
                   'Churn Reason','Churn Category','Customer ID'],axis =1, inplace = True)
```

```python
df = df.loc[df["Customer Status"] != 'Joined']
```

```python
# Define the numerical and categorical features
numerical_features = df.select_dtypes(include='number').columns.tolist()

# categorical_features = X.select_dtypes(include='object').drop('Customer Status',axis = 1).columns.tolist()
categorical_features = df.select_dtypes(include='object').columns.tolist()
```

```python
num_imputer = KNNImputer(n_neighbors=2)

# Iterate over each categorical feature in the DataFrame
for feature in numerical_features:
    df[feature] = num_imputer.fit_transform(df[[feature]])
```

```python
# drop rows with missing values
df.dropna(inplace=True)
```

```python
#choose best features 
X=df[['Married', 'Number of Dependents', 'Number of Referrals',
       'Tenure in Months', 'Offer', 'Internet Service', 'Online Security',
       'Online Backup', 'Device Protection Plan', 'Premium Tech Support',
       'Contract', 'Paperless Billing', 'Total Charges',
       'Total Long Distance Charges', 'Total Revenue','Age']]
y=df['Customer Status']
```

```python
# Define the numerical and categorical features
numerical_features_X = X.select_dtypes(include='number').columns.tolist()

# categorical_features = X.select_dtypes(include='object').drop('Customer Status',axis = 1).columns.tolist()
categorical_features_X = X.select_dtypes(include='object').columns.tolist()
```

```python
num_imputer = KNNImputer(n_neighbors=2)

# Iterate over each categorical feature in the DataFrame
for feature in numerical_features_X:
    X[feature] = num_imputer.fit_transform(X[[feature]])
```

```python
for col in categorical_features_X:
    print(col,X[col].value_counts())
    print()
```

```python
label_encoder = LabelEncoder()

# Iterate over each categorical feature in the DataFrame
for feature in categorical_features_X:
    X[feature] = label_encoder.fit_transform(df[feature])
```

```python
for col in categorical_features_X:
    print(col,X[col].value_counts())
    print()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def custmer_state(Status):
    if  Status=='Stayed':
        return 0
    else:
        return 1
```

```python
y_train = y_train.apply(custmer_state)
y_test = y_test.apply(custmer_state)
```

```python
# Instantiate the StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the training data and transform it
X_train = scaler.fit_transform(X_train)

# Use the scaler to transform the test data
X_test = scaler.transform(X_test)
```

```python
# oversample the minority class using SMOTE
smote = SMOTE(sampling_strategy='auto')
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

**4. Model**

```python
Models = [
        
        ("SVM",      SVC()),     #Support Vector Machines
               
        ("kNN",      KNeighborsClassifier(n_neighbors = 3)),    #KNeighborsClassifier

        ("LR_model", LogisticRegression(random_state=42,n_jobs=-1)),   #Logistic Regression model

        ("DT_model", DecisionTreeClassifier(random_state=42)),    #Decision tree model

        ("RF_model", RandomForestClassifier(random_state=42, n_jobs=-1)),   #Random Forest model

        ("GradientBoosting",GradientBoostingClassifier(max_depth=2,     #GradientBoosting model
                                                      n_estimators=100))
        ]
```

```python
accuracies = {}
models = {}
model = Models
for name,model in Models:
    model.fit(X_resampled, y_resampled)
    y_pre = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pre)*100
    accuracies[name] = accuracy
    print("{} Accuracy Score : {:.3f}%".format(name,accuracy))
```

```python
def evaluate_classification(y_test, y_pred):
    """
    This function takes in two arrays: y_true (true labels) and y_pred (predicted labels).
    It returns a dictionary with various evaluation metrics.
    """
    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    matrix = confusion_matrix(y_test, y_pred)
    
    print('accuracy: ',accuracy)
    print('confusion matrix: ')
    print(matrix)
```

```python
RF = RandomForestClassifier(
    random_state=42,
    n_jobs=-1,
    n_estimators=200,
    min_samples_split=10,
    min_samples_leaf=1,
    max_depth=None
)

RF.fit(X_resampled, y_resampled)
# Preprocess the test data using the pipeline and make predictions
y_pred = RF.predict(X_test)

evaluate_classification(y_test, y_pred)
```
